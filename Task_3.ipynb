# Visualizing the data


```python
import seaborn as sns
sns.pairplot(data)
```

```python
data.corr()
plt.figure(figsize = (10,5))
sns.heatmap(data.corr(),annot = True)
plt.show()
```

```python
#piechart
plt.figure(figsize = (6,6))
textprops = {"fontsize":13}
plt.title('Category wise analysis')
plt.pie(data['Category'].value_counts(),labels=data['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```python
#bargraph

plt.figure(figsize=(11,17))
data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
data.groupby('Sub-Category')['Profit','Sales'].agg(sum).plot.bar()
plt.title('Top Sub-Categories Sales vs Profit',fontsize = 15)
```

From the above graph,the Profit of Copiers are high,and the Sales of Chairs & Phones are high among the others.

```python
# A more detailed view
       
plt.figure(figsize=(8,10))
statwise = data.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statwise.plot.barh()                      # h for horizontal 
```

```python
# Region wise analysis

plt.figure(figsize = (7,7))
plt.title('Region')
plt.pie(data['Region'].value_counts(),labels=data['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

**Conclusion of Data Set -**

```python
plt.style.use('seaborn')
data.plot(kind = "scatter",figsize = (10,5), x = "Sales", y = "Profit",c = "Discount", s = 20,fontsize = 15, colormap = "ocean")
plt.ylabel('Total Profits',fontsize = 15)
plt.title("Interdependency of sales,Profits and Discounts",fontsize = 15)
plt.show()
```

**From the graph the one thing is clear that if we increase discount value then our sales will be increase but comparativly our profit goes down.**
